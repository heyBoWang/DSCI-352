```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
import seaborn as sns
```

```python
# Create a dataframe with columns named 'x' and 'y' filled with random numbers
df = pd.DataFrame({"x": np.random.normal(size=50), "y": np.random.normal(size=50)})

# Create an ordinary least squares regression on the data and show summary statistics.
# This formula regresses column 'y' on 'x'
ro = sm.OLS.from_formula("y ~ x", df)
res = ro.fit()
print(res.summary())

# Plot the data
sns.regplot('x', 'y', df)
plt.show()
```

```python

DATA_ROOT = '/Users/wanghaobo/Documents/ALL CSV FILES - 2nd Edition/'

advertising = pd.read_csv(DATA_ROOT+'Advertising.csv')
advertising.info()
advertising.describe()
```

```python
# Regress column 'sales' on 'TV'
ro = sm.OLS.from_formula("sales ~ TV", advertising)
res = ro.fit()
print(res.summary())

# Plot the data
sns.regplot('TV', 'sales', advertising)
plt.show()
```

```python
# Regress column 'sales' on 'radio'
ro = sm.OLS.from_formula("sales ~ radio", advertising)
res = ro.fit()
print(res.summary())

# Plot the data
sns.regplot('radio', 'sales', advertising)
plt.show()
```

```python
# Regress column 'sales' on 'newspaper'
ro = sm.OLS.from_formula("sales ~ newspaper", advertising)
res = ro.fit()
print(res.summary())

# Plot the data
sns.regplot('newspaper', 'sales', advertising)
plt.show()
```

# Results and Comparison


As shown in the three regression result tables above, the simple regression line using 'TV' has an r-squared value of 0.612, which is the highest among the three. This indicates that around 61.2% of the variability in sales can be explained by TV advertisements. The regression using 'radio' has an r-squared value of 0.332 and 'newspaper' has an r-squared of 0.052. These values are significantly lower than that of 'TV' regression, indicating a less predictive power.
The small p-values of the regression coefficient in the three regression lines are 0 ('TV'), 0 ('radio'), and 0.001 ('newspaper'), which indicate the significance of the explanatory variable. At a 95% confidence level, we can reject the null hypothesis that the regression coefficient is zero (i.e., no effect) in each of the three cases. The confidence intervals indicate the range of values we expect the coefficient to fall into with a 95% confidence. For example, we know with 95% confidence that the regression coefficient of 'TV' advertisement is between 0.042 and 0.053. The interpretation is that for every $1000 spent on TV advertisements, we can expect an increase in 4.2 to 5.3 units in sales with a 95% confidence.
